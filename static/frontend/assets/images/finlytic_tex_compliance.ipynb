```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Set a random seed for reproducibility
np.random.seed(42)

# Generate sample data
num_entries = 100
start_date = datetime(2024, 1, 1)

# Define expense categories relevant to a small business in Nepal
categories = [
    "Rent", "Utilities", "Salaries", "Inventory", "Marketing",
    "Office Supplies", "Transportation", "Maintenance", "Insurance",
    "Professional Fees", "Taxes", "Miscellaneous"
]

# Generate random data
dates = [start_date + timedelta(days=i) for i in range(num_entries)]
categories_list = np.random.choice(categories, num_entries)
amounts = np.random.randint(1000, 100000, num_entries)  # Amounts in Nepalese Rupees (NPR)

# Create DataFrame
df = pd.DataFrame({
    'Date': dates,
    'Category': categories_list,
    'Amount': amounts,
    'Description': [f"Expense for {cat}" for cat in categories_list]
})

# Sort by date
df = df.sort_values('Date')

# Add some specific entries to make it more realistic
specific_entries = [
    {'Date': datetime(2024, 1, 1), 'Category': 'Rent', 'Amount': 50000, 'Description': 'Monthly office rent'},
    {'Date': datetime(2024, 1, 5), 'Category': 'Utilities', 'Amount': 15000, 'Description': 'Electricity and water bill'},
    {'Date': datetime(2024, 1, 31), 'Category': 'Salaries', 'Amount': 200000, 'Description': 'Staff salaries for January'},
    {'Date': datetime(2024, 2, 15), 'Category': 'Taxes', 'Amount': 75000, 'Description': 'Quarterly VAT payment'},
    {'Date': datetime(2024, 3, 1), 'Category': 'Insurance', 'Amount': 30000, 'Description': 'Annual business insurance premium'}
]

df = pd.concat([df, pd.DataFrame(specific_entries)], ignore_index=True)
df = df.sort_values('Date').reset_index(drop=True)

# Calculate some summary statistics
total_expenses = df['Amount'].sum()
expenses_by_category = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)
highest_expense_category = expenses_by_category.index[0]
highest_expense_amount = expenses_by_category.iloc[0]

# Print summary
print(f"Total Expenses: NPR {total_expenses:,.2f}")
print(f"Highest Expense Category: {highest_expense_category} (NPR {highest_expense_amount:,.2f})")
print("\nTop 5 Expense Categories:")
print(expenses_by_category.head().to_string())

# Save to Excel
excel_file = 'expenses.xlsx'
df.to_excel(excel_file, index=False, engine='openpyxl')
print(f"\nSample expense data saved to {excel_file}")

# Optionally, create a pie chart of expenses by category
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
plt.pie(expenses_by_category, labels=expenses_by_category.index, autopct='%1.1f%%')
plt.title('Expenses by Category')
plt.axis('equal')
plt.savefig('expenses_pie_chart.png')
print("Pie chart of expenses saved as expenses_pie_chart.png")
```

```python
import pandas as pd
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer, Trainer, TrainingArguments, GPT2Config
from datasets import Dataset
from sklearn.model_selection import train_test_split
import os
import logging
import numpy as np
from tqdm import tqdm

# Set up logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Part 1: Data Preparation and Model Training

# Expanded dataset with more diverse tax-related questions and answers specific to Nepal
data = {
    'input': [
        "What are the current income tax brackets in Nepal?",
        "How do I calculate my taxable income as a freelancer in Nepal?",
        "What are the tax implications of selling property in Nepal?",
        "Can you explain the tax benefits for small businesses in Nepal?",
        "How does Nepal's tax system handle foreign income?",
        "What are the tax deduction limits for charitable donations in Nepal?",
        "How do I report rental income on my tax return in Nepal?",
        "What are the tax implications of investing in the stock market in Nepal?",
        "How does Nepal's tax system treat retirement contributions?",
        "What are the tax obligations for e-commerce businesses in Nepal?",
        "What is the VAT rate in Nepal and how is it applied?",
        "How are capital gains taxed in Nepal?",
        "What are the tax implications for expatriates working in Nepal?",
        "How does Nepal's tax system handle agricultural income?",
        "What are the tax incentives for foreign investors in Nepal?",
        # Add more questions and answers here to expand the dataset
    ],
    'response': [
        "Nepal's income tax brackets for individuals as of the fiscal year 2021/2022 are as follows: 1) First NPR 400,000: 1%, 2) Next NPR 100,000: 10%, 3) Next NPR 200,000: 20%, 4) Next NPR 1,300,000: 30%, 5) Above NPR 2,000,000: 36%. These rates may be subject to change, so it's advisable to check the latest information from the Inland Revenue Department of Nepal.",
        "As a freelancer in Nepal, calculate your taxable income by: 1) Summing up all your earnings, 2) Subtracting allowable expenses (e.g., equipment, internet, workspace costs), 3) Applying any applicable deductions or exemptions. Keep detailed records of income and expenses. You may need to make advance tax payments quarterly. Consult with a Nepalese tax professional for personalized advice.",
        "When selling property in Nepal, you're subject to capital gains tax. For land and buildings held more than 5 years, the tax rate is 2.5% of the sales proceeds or 5% of the capital gain, whichever is lower. For properties held less than 5 years, the rate is 5% of the sales proceeds or 10% of the capital gain, whichever is lower. Certain exemptions may apply for primary residences.",
        "Small businesses in Nepal can benefit from several tax incentives: 1) Reduced corporate tax rates (20% instead of 25% for some industries), 2) Tax holidays for certain sectors and regions, 3) Deductions for job creation, 4) Simplified bookkeeping requirements for very small businesses. The specific benefits depend on factors like industry, location, and annual turnover.",
        "Nepal taxes its residents on their worldwide income. Foreign income is generally taxable, but you may be eligible for foreign tax credits to avoid double taxation if Nepal has a tax treaty with the source country. Non-residents are taxed only on their Nepal-sourced income. It's important to report all foreign income and consult with a Nepalese tax expert for proper compliance.",
        "In Nepal, individuals can claim tax deductions for charitable donations made to approved institutions. The deduction is limited to either NPR 100,000 or 5% of your adjusted taxable income, whichever is lower. To claim the deduction, you must have proper documentation of the donations and ensure they were made to organizations recognized by the Inland Revenue Department.",
        "To report rental income in Nepal: 1) Calculate your gross rental income for the year, 2) Subtract allowable expenses (e.g., maintenance, property taxes, insurance), 3) Report the net rental income on your tax return. The tax rate depends on your total taxable income. Keep detailed records of all rental income and expenses. If your annual rental income exceeds NPR 2 million, you must register for VAT.",
        "Stock market investments in Nepal are subject to capital gains tax. For listed companies, the rate is 5% for individual investors and 10% for institutional investors. Dividends are taxed at 5% for both individuals and institutions. Long-term investments (held over a year) may have preferential treatment. Keep detailed records of all transactions for accurate reporting.",
        "Retirement contributions in Nepal can offer tax benefits. Contributions to approved retirement funds (like the Citizen Investment Trust or Employee Provident Fund) are tax-deductible up to certain limits. The current limit is one-third of your salary or NPR 300,000 per year, whichever is lower. Withdrawals after retirement are generally taxed at a lower rate.",
        "E-commerce businesses in Nepal have several tax obligations: 1) Income Tax on profits, 2) VAT registration if annual turnover exceeds NPR 5 million, 3) Social Security Fund contributions for employees, 4) Customs duties on imported goods. They must maintain proper books of accounts, issue invoices for all sales, and file regular tax returns. Digital transactions may be subject to additional reporting requirements.",
        "The standard VAT rate in Nepal is 13%. It applies to most goods and services, but there are some exemptions and zero-rated items. Businesses with an annual turnover exceeding NPR 5 million must register for VAT. VAT is collected at each stage of the supply chain and is ultimately borne by the final consumer.",
        "Capital gains in Nepal are taxed differently based on the type of asset and holding period. For example, gains from the sale of land or buildings are taxed at 2.5% to 5% of the sales value, depending on the holding period. Gains from the sale of shares in listed companies are taxed at 5% for individuals and 10% for institutions.",
        "Expatriates working in Nepal are generally considered tax residents if they stay in Nepal for 183 days or more in a fiscal year. They are taxed on their Nepal-sourced income at the same progressive rates as Nepalese citizens. Some expatriates may be eligible for tax exemptions or special provisions based on bilateral agreements between Nepal and their home countries.",
        "Agricultural income in Nepal is generally tax-exempt for individuals. However, large-scale commercial farming operations may be subject to income tax. It's important to maintain proper records and consult with a tax professional to determine if your agricultural activities qualify for tax exemption.",
        "Nepal offers various tax incentives to attract foreign investment, including: 1) Tax holidays for certain industries and regions, 2) Reduced corporate tax rates for export-oriented businesses, 3) Exemptions on customs duties and VAT for imported machinery and equipment, 4) Deductions for research and development expenses. The specific incentives depend on the sector and location of the investment.",
        # Add more responses here to match the additional questions
    ]
}

# Load data into a pandas DataFrame
df = pd.DataFrame(data)

# Split the data into training and validation sets
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)

# Tokenizer and Model
model_name = "gpt2-medium"  # Using a larger model
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
config = GPT2Config.from_pretrained(model_name)
config.num_labels = 1  # For regression task
model = GPT2LMHeadModel.from_pretrained(model_name, config=config)

# Set the padding token and padding side
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "left"

# Tokenize the dataset
def preprocess_function(examples):
    # Combine input and response
    full_text = [f"<|startoftext|>Question: {q}\nAnswer: {a}<|endoftext|>" for q, a in zip(examples['input'], examples['response'])]

    # Tokenize with padding and truncation
    tokenized = tokenizer(full_text, padding="max_length", truncation=True, max_length=512, return_tensors="pt")

    # Prepare labels (same as input_ids for causal language modeling)
    tokenized['labels'] = tokenized['input_ids'].clone()

    # Convert tensors to lists for Dataset compatibility
    for key in tokenized.keys():
        tokenized[key] = tokenized[key].tolist()

    logging.info(f"Tokenized shape: {len(tokenized['input_ids'])}, {len(tokenized['input_ids'][0])}")

    return tokenized

# Create Hugging Face Datasets
train_dataset = Dataset.from_pandas(train_df)
val_dataset = Dataset.from_pandas(val_df)

tokenized_train = train_dataset.map(preprocess_function, batched=True, remove_columns=train_dataset.column_names)
tokenized_val = val_dataset.map(preprocess_function, batched=True, remove_columns=val_dataset.column_names)

# Fine-tuning setup
training_args = TrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    num_train_epochs=20,
    logging_dir="./logs",
    logging_steps=10,
    evaluation_strategy="steps",
    eval_steps=50,
    save_steps=50,
    warmup_steps=500,
    learning_rate=5e-5,
    weight_decay=0.01,
    fp16=True,
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss",
    greater_is_better=False,
)

# Define the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_val,
)

# Fine-tune the model
logging.info("Starting model training...")
trainer.train()
logging.info("Model training completed.")

# Save the fine-tuned model
model_save_path = "./nepal_tax_compliance_gpt2"
model.save_pretrained(model_save_path)
tokenizer.save_pretrained(model_save_path)
logging.info(f"Model saved to {model_save_path}")

# Part 2: Expense Analysis and Real-time Advice Generation

def analyze_expenses(file_path):
    try:
        df = pd.read_excel(file_path)
        # Assuming the Excel file has 'Category' and 'Amount' columns
        categorized_expenses = df.groupby('Category')['Amount'].sum().to_dict()
        total_expenses = sum(categorized_expenses.values())
        return total_expenses, categorized_expenses
    except Exception as e:
        logging.error(f"Error analyzing expenses: {str(e)}")
        return None, None

def generate_tax_advice(query, expense_file_path, model, tokenizer):
    try:
        # Analyze expenses
        total_expenses, categorized_expenses = analyze_expenses(expense_file_path)

        if total_expenses is None or categorized_expenses is None:
            return "Error: Unable to analyze expenses."

        # Prepare input for the model
        expense_summary = ", ".join([f"{category}: {amount}" for category, amount in categorized_expenses.items()])
        full_query = f"<|startoftext|>Question: {query}\nContext: The user's expenses in Nepal are as follows: Total expenses: {total_expenses} NPR, Categorized expenses: {expense_summary}\nAnswer:"

        # Tokenize input
        inputs = tokenizer.encode(full_query, return_tensors="pt", add_special_tokens=True)

        # Generate response
        with torch.no_grad():
            outputs = model.generate(
                inputs,
                max_length=512,
                num_return_sequences=1,
                temperature=0.7,
                top_k=50,
                top_p=0.95,
                do_sample=True,
                no_repeat_ngram_size=2,
                pad_token_id=tokenizer.eos_token_id
            )

        response = tokenizer.decode(outputs[0], skip_special_tokens=True)
        return response.split("Answer:", 1)[-1].strip()
    except Exception as e:
        logging.error(f"Error generating tax advice: {str(e)}")
        return "Error: Unable to generate tax advice."

# Load the fine-tuned model and tokenizer
logging.info("Loading fine-tuned model and tokenizer...")
model_path = "./nepal_tax_compliance_gpt2"
model = GPT2LMHeadModel.from_pretrained(model_path)
tokenizer = GPT2Tokenizer.from_pretrained(model_path)
logging.info("Model and tokenizer loaded successfully.")

# Example usage
expense_file_path = "expenses.xlsx"  # Path to the local expense file
user_query = "What tax deductions can I claim based on my expenses in Nepal?"

logging.info("Generating tax advice...")
advice = generate_tax_advice(user_query, expense_file_path, model, tokenizer)
print("Tax Advice:")
print(advice)

# Post-processing function to improve response quality
def post_process_response(response):
    # Remove any mentions of non-Nepalese tax systems
    non_nepal_keywords = ["US", "UK", "India", "China"]
    for keyword in non_nepal_keywords:
        response = response.replace(keyword, "Nepal")

    # Ensure the response ends with a complete sentence
    if not response.endswith((".", "!", "?")):
        response = ". ".join(response.split(".")[:-1]) + "."

    return response

# For web application integration
def get_tax_advice(query, expense_file_path):
    raw_advice = generate_tax_advice(query, expense_file_path, model, tokenizer)
    return post_process_response(raw_advice)

if __name__ == "__main__":
    # You can add more example queries here
    print("\nAdditional example queries:")
    print("\nQuery: What are my potential tax liabilities in Nepal?")
    print(get_tax_advice("What are my potential tax liabilities in Nepal?", expense_file_path))

    print("\nQuery: How can I optimize my tax strategy based on these expenses in Nepal?")
    print(get_tax_advice("How can I optimize my tax strategy based on these expenses in Nepal?", expense_file_path))
```
